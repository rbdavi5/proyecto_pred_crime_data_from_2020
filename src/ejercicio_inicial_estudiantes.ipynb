## Ejercicio 6: Crear un boxplot de las edades de las víctimas

Crea un boxplot que muestre la distribución de las edades de las víctimas. Filtra primero los valores para incluir solo edades entre 0 y 100 años.

## Ejercicio 8: Boxplot comparativo

Crea un boxplot que compare las edades de las víctimas según su sexo. Filtra primero para incluir solo los valores 'M' y 'F' en la columna 'Vict Sex', y edades entre 0 y 100.

```python
# Calcular IQR
q1 = df['Vict Age'].quantile(0.25)
q3 = df['Vict Age'].quantile(0.75)
iqr = q3 - q1
print(f"Q1: {q1}, Q3: {q3}, IQR: {iqr}")

# Porcentaje de nulos
total = len(df)
print(f"\nTotal de registros: {total}")
print(f"\nPorcentaje de nulos en columnas clave:")
print(f"Vict Sex: {(df['Vict Sex'].isnull().sum() / total * 100):.2f}%")
print(f"Vict Age: {(df['Vict Age'].isnull().sum() / total * 100):.2f}%")
print(f"Weapon Desc: {(df['Weapon Desc'].isnull().sum() / total * 100):.2f}%")
```
